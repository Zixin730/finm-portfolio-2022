```python
import numpy as np
import statsmodels.api as sm
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd
```

```python
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

# Part 1
# Please do not modify this seed
np.random.seed(6996)

# error term
epsilon_vec = np.random.normal(0, 1, 500)  # to be completed
# X_matrix or regressors or predictiors
X_mat = np.random.normal(0, 2, 500 * 500)  # to be completed
X_mat = X_mat.reshape((500, 500))
# Slope
slope_vec = np.random.uniform(1, 5, 500)  # to be completed
# Simulate Ys
Y_mat = np.array(list(map(lambda z: list(
    1 + np.sum(X_mat[:, :z] * slope_vec[:z], axis=1) + epsilon_vec), range(2, 501)))).T
Y_mat
```

```python
# Create linear regression object
n = 388
X = sm.add_constant(X_mat[:, :n+1])
regr = sm.OLS(Y_mat[:, n], X).fit()  # To be completedv
regr.summary()
```

```python
# create the function add_reg which will add a number of regressor
def add_reg(num, level=0.05):
    if num < 2:
        print('not enough regressor selected')
        return
    else:

        # Regressor Matrix ＋ Constant
        X = X_mat[:, :num]  # To be completed
        X = sm.add_constant(X)
        # Dependent Variable Vector
        y = Y_mat[:, num-2]  # To be completed
        ols_reg = sm.OLS(y, X).fit()  # To be completed
        # R square
        r2 = ols_reg.rsquared  # To be completed
        # confidence interval (choose alpha = level by default = 0.05)
        intv = ols_reg.conf_int(alpha=0.05, cols=None)  # To be completed
        return r2, intv
     # (store the rsquared and the confidence interval min max here)


def test_part1():
    print(X_mat)


def test_part1_1():
    print(Y_mat.shape)


def test_part1_2():
    print(Y_mat)


def test_part2():
    print(regr.params)


def test_part2_1():
    print(regr.pvalues)


def test_part3():
    result = list(map(add_reg, range(2, 501, 100)))
    print(result)
```
